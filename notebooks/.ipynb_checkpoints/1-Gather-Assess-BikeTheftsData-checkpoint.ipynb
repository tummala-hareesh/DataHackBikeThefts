## Gather data

Bike theft data is openly available on [Toronto Public Service - PUBLIC SAFETY DATA PORTAL](https://data.torontopolice.on.ca/datasets/698717de582845458159a73ad4a2a8e0_0). TPS provides two possible ways to extract data. First method is everyday method of downloading relevant files and reading data from those files. Second method is by using an Restful API. Here, we use 1st method to extract and load bike theft data.
